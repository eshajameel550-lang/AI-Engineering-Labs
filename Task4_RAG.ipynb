```python
import os
from dotenv import load_dotenv
from google import genai
```

```python
load_dotenv()

api_key = os.getenv("GOOGLE_API_KEY")

client = genai.Client(api_key=api_key)

MODEL_NAME = "gemini-2.5-flash"

print("Gemini API Connected Successfully!")
```

```python
documents = [
    {
        "content": "Machine Learning is a subset of Artificial Intelligence that learns from data.",
        "category": "AI"
    },
    {
        "content": "Deep Learning uses neural networks with multiple hidden layers.",
        "category": "AI"
    },
    {
        "content": "Employee onboarding includes training and documentation.",
        "category": "HR"
    },
    {
        "content": "HR policies include leave management and employee benefits.",
        "category": "HR"
    }
]

print(documents)
```

```python
hr_docs = [doc for doc in documents if doc["category"] == "HR"]

print("HR Documents")

for doc in hr_docs:
    print(doc)
```

```python
context = "\n".join([doc["content"] for doc in hr_docs])

prompt = f"""
Answer only using the following context.

Context:
{context}

Question:
What are HR policies?
"""

response = client.models.generate_content(
    model=MODEL_NAME,
    contents=prompt
)

print(response.text)
```

```python
print("Retrieved Documents:", len(hr_docs))

print("Categories:")

for doc in hr_docs:
    print(doc["category"])
```
